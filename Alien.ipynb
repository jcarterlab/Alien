## 1) Preparation

### Import libaries & read in data

```python
# imports libaries and reads in the Hofstede cultural dimensions data. 
import pandas as pd 
import numpy as np
import math as m

data = pd.read_excel('C:/Users/HUAWEI/Desktop/Exploratory data/Hofstede/Data.xls')
data.head(4)
```

### Check missing values

```python
# gets the na count for a single row. 
def get_row_nas(row):
    
    counter = 0
    for col in range(len(row)):
        if type(row.iloc[col]) == np.float64:
            if np.isnan(row.iloc[col]):
                counter = counter + 1
    return(counter)

# gets the na count for the entire data frame.
def get_data_nas(df):
    
    na_count = []
    for row in range(len(df)):
        na_count.append(get_row_nas(df.iloc[row]))
    
    return(na_count)
```

```python
# creates an na count column.
data['na_count'] = get_data_nas(data)

# summarizes the number of nas in the data frame.  
data.na_count.value_counts()
```

### Filter missing values

```python
# filters all rows with greater than 2 na values. 
filtered_df = data.loc[data.na_count<=2]
```

```python
# lists how many na values for each column.
filtered_df.isnull().sum()
```

### Fill missing values

```python
# 
pd.set_option('mode.chained_assignment', None) 

impute_var = []

for i in range(len(filtered_df)):
    impute_var.append(1 if (round(filtered_df.pdi.iloc[i]/10) >= 5) else 0)
    
def fill_missing_values(data, col): 
    
    data['impute_var'] = impute_var
    
    avg = data.groupby('impute_var')[col].mean()
    
    guess = []
    for i in range(len(data)):
        guess.append(avg[0] if (data['impute_var'].iloc[i]==0) else avg[1])
        
    for i in range(len(data)):
        data[col].iloc[i] = guess[i] if (np.isnan(data[col].iloc[i])) else data[col].iloc[i]
    
    data.drop('impute_var', axis=1, inplace=True)
    
    return(data)

fill_missing_values(filtered_df, 'ltowvs')
fill_missing_values(filtered_df, 'ivr')
```

```python
# lists how many na values for each column.
filtered_df.isnull().sum()
```

```python
filtered_df.drop('na_count', axis=1, inplace=True)
```

### Deal with outliers

```python
def check_outliters(col):
    mean = np.mean(col)
    sd = np.std(col)
    outliers = []
    
    for y in col:
        z_score = (y - mean) / sd
        if np.abs(z_score) > 2:
            outliers.append(col.name)
            
    return(outliers)

dims = renamed_df.columns.values[1:len(renamed_df.columns.values)]
            
for i in range(len(dims)):
    outliers = check_outliters(renamed_df.loc[:,dims[i]])
    
outliers
```

## 2) K-means clustering

```python
from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = renamed_df.iloc[:,1:renamed_df.shape[1]]
y = renamed_df.iloc[:,0:1]

x_std = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
pca_x = pca.fit_transform(x_std)

print('PCA retains ' + str(round(pca.explained_variance_ratio_[0], 2)*100) + '% & ' + str(round(pca.explained_variance_ratio_[1], 2)*100) + '% of the variance.')
```

```python
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', max_iter = 300, n_init = 100, random_state = 0)
    kmeans.fit(pca_x)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(1, 11), inertia)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# creates the clusters.
kmeans = KMeans(n_clusters=3, init = 'k-means++', max_iter = 300, n_init = 100, random_state = 0)
kmeans.fit(pca_x)
renamed_df['clusters'] = kmeans.fit_predict(pca_x)
```

```python
# imports libaries.
from scipy.spatial import ConvexHull
from scipy import interpolate
from adjustText import adjust_text
    
# creates the scatter plot.
fig, ax = plt.subplots(figsize=(8,6), constrained_layout=True)
ax.scatter(pca_x[:,1], pca_x[:,0], marker='.', c=renamed_df['clusters'], cmap='rainbow')

# assigns cluster colors. 
colors = ['#800080', '#008000', '#800000', '#00FFFF', '#000000']
renamed_df['colors'] = ''
for i in range(len(colors)):
    renamed_df.colors[renamed_df.clusters==i] = colors[i]
    
# creates the convexes. 
for i in renamed_df.clusters.unique():
    index = renamed_df[renamed_df.clusters==i].index
    df = pd.DataFrame()
    df['1'] = pca_x[index,1]
    df['2'] = pca_x[index,0]
    points = df.loc[:,['1', '2']].values
    
    hull = ConvexHull(points)
    x_hull = np.append(points[hull.vertices,0],points[hull.vertices,0][0])
    y_hull = np.append(points[hull.vertices,1],points[hull.vertices,1][0])

    dist = np.sqrt((x_hull[:-1] - x_hull[1:])**2 + (y_hull[:-1] - y_hull[1:])**2)
    dist_along = np.concatenate(([0], dist.cumsum()))
    spline, u = interpolate.splprep([x_hull, y_hull], u=dist_along, s=0, per=1)
    interp_d = np.linspace(dist_along[0], dist_along[-1], 100)
    interp_x, interp_y = interpolate.splev(interp_d, spline)
    plt.fill(interp_x, interp_y, '--', c=colors[i], alpha=0.1)
    
# assigns the labels. 
labels = renamed_df.country    
texts = []
for i in range(len(labels)):
    x = pca_x[i,1]
    y = pca_x[i,0]
    text = labels.iloc[i]
    texts.append(ax.text(x, y, text, fontsize=7))

adjust_text(texts,
            expand_points=(2.25, 2.25),
            arrowprops=dict(arrowstyle="->", lw=0.65),
            ax=fig.axes[0])
    
# visualizes the data. 
plt.title('Hofstede\'s Cultural Dimensions')
plt.xlabel('PCA var 2')
plt.ylabel('PCA var 1')
plt.xlim(-3,4)
plt.ylim(-3,3.5)
plt.show()
```
